```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def gaussian(t, mu, sig):
    return np.exp(-np.power(t - mu, 2.) / (2 * np.power(sig, 2.)))
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, sharey=True)

for axs in (ax1, ax2):
    axs.spines["left"].set_position(("data", 0))
    axs.spines["bottom"].set_position(("data", 0))
    
    axs.spines["top"].set_visible(False)
    axs.spines["right"].set_visible(False)
    
    plt.xticks([])
    plt.yticks([0, 1], ['', 'A'])
    
    axs.plot(1, 0, ">k", transform=axs.get_yaxis_transform(), clip_on=False)
    axs.plot(0, 1, "^k", transform=axs.get_xaxis_transform(), clip_on=False)
        
    plt.xlim(-2*np.pi, 20)
    plt.ylim(-0.1, 1.25)

    axs.text(20, -0.2, "$x$")
    axs.text(-1, 1.25, "$y$")
    
    axs.tick_params(axis='y', labelsize=14, direction='inout', length=12)

ax1.arrow(2, 0.5, 2, 0, 
          head_width=0.05, head_length=0.25, 
          fc='g', ec='g')
ax2.arrow(14, 0.5, 2, 0, 
          head_width=0.05, head_length=0.25, 
          fc='g', ec='g')

ax1.text(3, 0.6, "$V$", fontsize=11)
ax2.text(15, 0.6, "$V$", fontsize=11)

ax1.text(-5, 0.2, "$t = 0$", fontsize=12)
ax2.text(-5, 0.2, "$t = t_{1}$", fontsize=12)

bbox=dict(fc="white", ec="none")

ax1.annotate("", xy=(-1.2*np.pi, -0.1), xytext=(1.2*np.pi, -0.1), textcoords=ax1.transData, arrowprops=dict(arrowstyle='<->'))
ax1.annotate("", xy=(-1.2*np.pi, -0.1), xytext=(1.2*np.pi, -0.1), textcoords=ax1.transData, arrowprops=dict(arrowstyle='|-|'))
ax1.text(0, -0.175, r"$\lambda$", ha="center", va="center", bbox=bbox, fontsize=12)

ax2.annotate("", xy=(-0.15, -0.1), xytext=(12, -0.1), textcoords=ax2.transData, arrowprops=dict(arrowstyle='<->'))
ax2.annotate("", xy=(-0.15, -0.1), xytext=(12, -0.1), textcoords=ax2.transData, arrowprops=dict(arrowstyle='|-|'))
ax2.text(6, -0.2, r"$ \Delta x = v\Delta t $", ha="center", va="center", bbox=bbox, fontsize=12)

t_values1 = np.linspace(-1.2*np.pi, 1.2*np.pi, 500)
ax1.plot(t_values1, gaussian(t_values1, 0, 1))

t_values2 = t_values1 + 12
ax2.plot(t_values2, gaussian(t_values1, 0, 1))

plt.subplots_adjust(hspace=0.25)

# uncomment this line to save the plot
# plt.savefig('unit1-1_01.jpg', dpi=300, bbox_inches='tight')

plt.show()
```
