```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2.5, 2.5, 4000)
P = (np.sin(np.pi * x) / (x))**2
single = P * 2
double = P * (1 + np.cos(2 * np.pi * 2.5 * x))

fig = plt.figure()
ax = fig.add_subplot()

ax.spines["left"].set_position(("data", 0))
ax.spines["bottom"].set_position(("data", 0))

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.xticks([])
plt.yticks([])

plt.plot(x, double, 'red')
plt.plot(x, single, 'gray')

# plt.savefig("lecture3-2_pattern.jpg", dpi=300, bbox_inches='tight')
```
